````python
def pipelining(df, col_name: str, value: int):
    df[col_name] = value
    df1 = df[df['Survived'] == 1]
    df1['Age'] = df1['Age'].fillna(df1['Age'].mean())

    return df1

df_titanic.pipe(pipelining, 'Fare_Per_Age', df_titanic['Fare'] / df_titanic['Age'])
````

````python
def pipe_flight(df, col_name: str, value: float):
    df1 = df[df["DepDelay"]>30]
    
    df1[col_name] = value
    
flights.pipe(pipe_flight, 'Delay_Per_Hour', (flights['DepDelay']) / (flights['ArrTime']))
````
